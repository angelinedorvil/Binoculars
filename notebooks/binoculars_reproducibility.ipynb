```python
import json, os, pandas as pd, matplotlib.pyplot as plt
from pathlib import Path
import glob
from sklearn.metrics import roc_curve, auc, f1_score, precision_recall_curve
import numpy as np

base = Path("/content/Binoculars/results")

# --- 1. Get folders sorted by modified time ---
folders = sorted(glob.glob(str(base / "*")), key=os.path.getmtime)
if not folders:
    raise FileNotFoundError("⚠️ No results found in /content/Binoculars/results")

print(f"📁 Found {len(folders)} result folders:")
for f in folders:
    print(" ├─", Path(f).name)

metrics_summary = []
plt.figure(figsize=(7, 6))

for folder in folders:
    folder = Path(folder)
    name = folder.name  # Full folder name like HK-News-ANP-Aya101-1024-tokens

    # --- 2. Safely load experiment files ---
    details_file = folder / "experiments_details.json"
    preds_file = folder / "score_df.csv"

    # Check existence explicitly
    if not details_file.exists() or not preds_file.exists():
        print(f"⚠ Skipping {name}: missing details or preds file.")
        continue

    details = json.load(open(details_file, "r"))
    preds = pd.read_csv(preds_file)

    # --- 3. Extract dataset ID (strip model/token suffix) ---
    # Example: "HK-News-ANP-Aya101-1024-tokens" → "HK-News-ANP"
    dataset_id = name.split("-Aya")[0].split("-GPT")[0]

    y_true = preds["class"]
    y_score = -1 * preds["score"]  # lower score → more machine-like

    fpr, tpr, _ = roc_curve(y_true, y_score, pos_label=1)
    auc_value = auc(fpr, tpr)

    # --- Best F1 threshold ---
    precision, recall, thresholds = precision_recall_curve(y_true, y_score)
    f1_scores = 2 * (precision * recall) / (precision + recall + 1e-9)
    best_f1_idx = np.nanargmax(f1_scores)
    best_f1 = f1_scores[best_f1_idx]
    best_thresh = thresholds[min(best_f1_idx, len(thresholds) - 1)]
    tpr_at_001 = np.interp(0.01, fpr, tpr)  # TPR @ FPR 1%

    metrics_summary.append({
        "Dataset": dataset_id,
        "Model-Run": name,
        "AUC": auc_value,
        "F1": best_f1,
        "TPR@FPR=0.01": tpr_at_001,
        "Best Threshold": best_thresh,
    })

    plt.plot(fpr, tpr, lw=2, label=f"{dataset_id} (AUC={auc_value:.3f})")

# --- 4. Plot and display ---
plt.plot([0,1], [0,1], "k--", lw=1, label="Random baseline")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Binoculars ROC Curves – Reproduced (F1-optimized)")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

df = pd.DataFrame(metrics_summary)
display(df)

# --- Paper references ---
paper_ref = pd.DataFrame([
    {"Dataset": "HK-News-ANP", "AUC": 0.9992, "F1": 0.9821, "TPR@FPR=0.01": 0.978},
    {"Dataset": "HK-News-AP", "AUC": 0.9992, "F1": 0.9821, "TPR@FPR=0.01": 0.978},
    {"Dataset": "HK-News-GNP", "AUC": 0.9992, "F1": 0.9821, "TPR@FPR=0.01": 0.978},
    {"Dataset": "HK-News-GP", "AUC": 0.9992, "F1": 0.9821, "TPR@FPR=0.01": 0.978},
    {"Dataset": "HK-essays-ANP", "AUC": 0.9985, "F1": 0.9813, "TPR@FPR=0.01": 0.965},
    {"Dataset": "HK-essays-AP", "AUC": 0.9985, "F1": 0.9813, "TPR@FPR=0.01": 0.965},
    {"Dataset": "HK-essays-GNP", "AUC": 0.9985, "F1": 0.9813, "TPR@FPR=0.01": 0.965},
    {"Dataset": "HK-essays-GP", "AUC": 0.9985, "F1": 0.9813, "TPR@FPR=0.01": 0.965},
    {"Dataset": "HK-blogs-ANP", "AUC": 0.9959, "F1": 0.9794, "TPR@FPR=0.01": 0.942},
    {"Dataset": "HK-blogs-AP", "AUC": 0.9959, "F1": 0.9794, "TPR@FPR=0.01": 0.942},
    {"Dataset": "HK-blogs-GNP", "AUC": 0.9959, "F1": 0.9794, "TPR@FPR=0.01": 0.942},
    {"Dataset": "HK-blogs-GP", "AUC": 0.9959, "F1": 0.9794, "TPR@FPR=0.01": 0.942},
])

merged = df.merge(paper_ref, on="Dataset", suffixes=("_yours", "_paper"))
merged["ΔAUC"] = merged["AUC_yours"] - merged["AUC_paper"]
merged["ΔF1"] = merged["F1_yours"] - merged["F1_paper"]
merged["ΔTPR@FPR=0.01"] = merged["TPR@FPR=0.01_yours"] - merged["TPR@FPR=0.01_paper"]

print("\n=== 📈 Differences (Your Run – Paper) ===")
display(merged[["Dataset", "Model-Run", "ΔAUC", "ΔF1", "ΔTPR@FPR=0.01"]])

# ============================================================
# SAVE OUTPUTS
# ============================================================

summary_dir = base / "summary_outputs"
summary_dir.mkdir(exist_ok=True)

# 1️ Save the ROC plot
plot_path = summary_dir / "binoculars_roc_curves.png"
plt.savefig(plot_path, dpi=300, bbox_inches="tight")
print(f"ROC curve plot saved → {plot_path}")

# 2️ Save full metrics summary table
metrics_csv_path = summary_dir / "metrics_summary.csv"
df.to_csv(metrics_csv_path, index=False)
print(f"Metrics summary saved → {metrics_csv_path}")

# 3️ Save paper-difference comparison
comparison_csv_path = summary_dir / "paper_comparison_metrics.csv"
merged.to_csv(comparison_csv_path, index=False)
print(f"Paper comparison table saved → {comparison_csv_path}")

# 4️ Create & save grouped summary by model/prompt setting
# Extract model and mode (GPT vs Aya / prompt vs no prompt)
def extract_model_info(name):
    # Example: "HK-News-ANP-Aya101-1024-tokens"
    parts = name.split("-")
    dataset = "-".join(parts[:2])     # HK-News
    mode = parts[2]                   # ANP, AP, GNP, GP
    return dataset, mode

df_grouped = df.copy()
df_grouped[["DatasetID", "Mode"]] = df_grouped["Model-Run"].apply(
    lambda x: pd.Series(extract_model_info(x))
)

# Pivot table: rows = Dataset, columns = Mode (ANP, AP, GNP, GP), values = metrics
metrics_pivot = df_grouped.pivot_table(
    index="DatasetID",
    columns="Mode",
    values=["AUC", "F1", "TPR@FPR=0.01"],
    aggfunc="first"
)

grouped_csv_path = summary_dir / "grouped_metrics_table.csv"
metrics_pivot.to_csv(grouped_csv_path)
print(f"Grouped metrics table saved → {grouped_csv_path}")

print("\nAll outputs saved successfully!")
```

```python
import os
import json
import torch
import hashlib

from transformers import AutoTokenizer, AutoModelForCausalLM
from concurrent.futures import ProcessPoolExecutor
from openai import OpenAI

# ============================================================
#  CONFIG
# ============================================================

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

# <-- Put actual key here or set in environment -->
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY", "key")

client = OpenAI(api_key=OPENAI_API_KEY)

# Input chunked files
INPUT_FILES = {
    "news":   "/content/Binoculars/extension/chunked_news.jsonl",
    "essays": "/content/Binoculars/extension/chunked_essays.jsonl",
    "blogs":  "/content/Binoculars/extension/chunked_blogs.jsonl",
}

# Output directory in Drive
OUTPUT_DIR = "/content/drive/MyDrive"

# Models
GEN_MODELS = {
    "aya101": {
        "type": "hf_chat",
        "name": "MaziyarPanahi/Mistral-7B-Instruct-Aya-101"
    },
    "gpt4o": {
        "type": "openai_chat",
        "name": "gpt-4o"
    }
}

# Shared generation settings
GEN_TEMP = 0.0           # 0.0 ~= greedy
GEN_TOP_P = 1.0
GEN_MAX_NEW_TOKENS = 512


# ============================================================
#  MODEL LOADING (lazy)
# ============================================================

loaded_hf_models = {}  # for Aya / HF models


def load_hf_chat_model(model_key):
    if model_key in loaded_hf_models:
        return loaded_hf_models[model_key]

    cfg = GEN_MODELS[model_key]
    model_name = cfg["name"]

    print(f"\n[HF] Loading model: {model_name}\n")

    tokenizer = AutoTokenizer.from_pretrained(model_name)

    model = AutoModelForCausalLM.from_pretrained(
        model_name,
        torch_dtype=torch.float16 if DEVICE == "cuda" else torch.float32
    ).to(DEVICE)

    loaded_hf_models[model_key] = (tokenizer, model)
    return loaded_hf_models[model_key]


# ============================================================
#  GENERATION HELPERS
# ============================================================

def build_with_prompt_text(text, min_chars):
    """English instruction + Haitian Creole article."""
    return (
        f"Continue or summarize this Haitian Creole news article. "
        f"Match the tone. Write at least {min_chars} characters.\n\n{text}"
    )


def generate_hf_chat(model_key, text, min_chars, mode):
    """
    mode: 'with_prompt' or 'no_prompt'
    Uses chat template for Aya-101.
    """
    tokenizer, model = load_hf_chat_model(model_key)

    if mode == "with_prompt":
        content = build_with_prompt_text(text, min_chars)
    else:  # no_prompt
        content = text

    messages = [
        {"role": "user", "content": content}
    ]

    inputs = tokenizer.apply_chat_template(
        messages,
        add_generation_prompt=True,
        return_tensors="pt"
    ).to(model.device)

    with torch.no_grad():
        output_ids = model.generate(
            inputs,
            max_new_tokens=GEN_MAX_NEW_TOKENS,
            do_sample=(GEN_TEMP > 0.0),
            temperature=GEN_TEMP,
            top_p=GEN_TOP_P
        )

    # Only decode newly generated tokens
    generated_ids = output_ids[0][inputs.shape[-1]:]
    generated = tokenizer.decode(generated_ids, skip_special_tokens=True)
    return generated.strip()


def generate_openai_chat(model_key, text, min_chars, mode):
    """
    GPT-4o via OpenAI chat completions.
    mode: 'with_prompt' or 'no_prompt'
    """
    model_name = GEN_MODELS[model_key]["name"]

    if mode == "with_prompt":
        content = build_with_prompt_text(text, min_chars)
    else:
        content = text

    response = client.chat.completions.create(
        model=model_name,
        messages=[
            {"role": "user", "content": content}
        ],
        temperature=GEN_TEMP,
        top_p=GEN_TOP_P,
        max_tokens=GEN_MAX_NEW_TOKENS,
    )

    out = response.choices[0].message.content
    return (out or "").strip()


def generate_for_model(model_key, text, min_chars, mode):
    """
    Unified wrapper.
    mode ∈ {'with_prompt', 'no_prompt'}
    """
    mtype = GEN_MODELS[model_key]["type"]

    if mtype == "hf_chat":
        return generate_hf_chat(model_key, text, min_chars, mode)
    elif mtype == "openai_chat":
        return generate_openai_chat(model_key, text, min_chars, mode)
    else:
        raise ValueError(f"Unknown model type: {mtype}")


# ============================================================
#  RESUME SUPPORT
# ============================================================

def row_key(entry):
    return entry.get("text", "").strip()[:800]

def load_existing_output(path):
    """
    Load existing JSONL into dict: key -> full entry.
    """
    if not os.path.exists(path):
        return {}

    data = {}
    with open(path, "r", encoding="utf-8") as f:
        for line in f:
            try:
                obj = json.loads(line)
            except json.JSONDecodeError:
                continue
            k = row_key(obj)
            data[k] = obj
    return data


# ============================================================
#  MAIN LOOP
# ============================================================

def process_file_for_model(model_key, dataset_name, input_path, output_dir):
    """
    For one model (aya101 or gpt4o) and one dataset (news / essays / blogs),
    read chunked JSONL and produce output JSONL in Drive.
    """
    os.makedirs(output_dir, exist_ok=True)
    out_path = os.path.join(output_dir, f"{dataset_name}_{model_key}.jsonl")
    tmp_path = out_path + ".tmp"

    print(f"\n===== DATASET: {dataset_name} | MODEL: {model_key} =====")
    print(f"Input:  {input_path}")
    print(f"Output: {out_path}")

    existing = load_existing_output(out_path)
    print(f"Loaded {len(existing)} existing rows for resume.\n")

    num_total = 0
    num_skipped = 0
    num_generated = 0

    with open(input_path, "r", encoding="utf-8") as f_in, \
         open(tmp_path, "w", encoding="utf-8") as f_out:

        for line in f_in:
            num_total += 1
            try:
                entry = json.loads(line)
            except json.JSONDecodeError:
                continue

            k = row_key(entry)
            text = entry.get("text", "")
            if not text.strip():
                continue

            min_chars = max(200, len(text) // 2)

            # look for existing entry to resume
            existing_entry = existing.get(k)
            if existing_entry is None:
                out_entry = dict(entry)
            else:
                out_entry = dict(existing_entry)

            with_key = f"{model_key}_with_prompt"
            nop_key = f"{model_key}_no_prompt"

            already_have_with = isinstance(out_entry.get(with_key), str) and out_entry[with_key].strip() != ""
            already_have_nop = isinstance(out_entry.get(nop_key), str) and out_entry[nop_key].strip() != ""

            # If both generations exist, just keep and skip
            if already_have_with and already_have_nop:
                num_skipped += 1
                f_out.write(json.dumps(out_entry, ensure_ascii=False) + "\n")
                continue

            # Otherwise, (re)generate missing ones
            try:
                if not already_have_nop:
                    gen_nop = generate_for_model(model_key, text, min_chars, mode="no_prompt")
                    out_entry[nop_key] = gen_nop

                if not already_have_with:
                    gen_with = generate_for_model(model_key, text, min_chars, mode="with_prompt")
                    out_entry[with_key] = gen_with

                num_generated += 1
                f_out.write(json.dumps(out_entry, ensure_ascii=False) + "\n")

                url_display = out_entry.get("url", "")[:80]
                print(f"✓ [{dataset_name}] {model_key} generated for: {url_display}")

            except Exception as e:
                # In case of API or HF error
                print(f"⚠ Error generating for key={k}: {e}")
                f_out.write(json.dumps(out_entry, ensure_ascii=False) + "\n")

    # Atomically replace
    os.replace(tmp_path, out_path)

    print(f"\nFinished dataset={dataset_name}, model={model_key}")
    print(f"Total rows:     {num_total}")
    print(f"Skipped (keep): {num_skipped}")
    print(f"Generated:      {num_generated}")
    print(f"Saved to:       {out_path}")


# ============================================================
#  RUN FOR ALL MODELS × DATASETS
# ============================================================

for model_key in GEN_MODELS.keys():
    for dataset_name, input_path in INPUT_FILES.items():
        process_file_for_model(model_key, dataset_name, input_path, OUTPUT_DIR)

# ============================================================
#  Alternative RUN FOR GPT4 MODEL (FASTER)
# ============================================================

# def all_jobs():
#     jobs = []
#     for model_key in GEN_MODELS.keys():
#         for dataset_name, input_path in INPUT_FILES.items():
#             jobs.append((model_key, dataset_name, input_path))
#     return jobs


# if __name__ == "__main__":
#     jobs = all_jobs()
#     with ProcessPoolExecutor(max_workers=6) as ex:
#         futures = []
#         for model_key, dataset_name, input_path in jobs:
#             futures.append(
#                 ex.submit(
#                     process_file_for_model,
#                     model_key,
#                     dataset_name,
#                     input_path,
#                     OUTPUT_DIR
#                 )
#             )
#         for f in futures:
#             f.result()  # ensure exceptions propagate
```
